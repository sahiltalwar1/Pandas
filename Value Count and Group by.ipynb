GROUP-BY QUESTIONS




What is the average price for each model?

```python
bdf=df.groupby(['model']).agg(
    avg_price = ('price','mean')).reset_index().head()
bdf
```

What is the average mileage for each model?


```python
bdf = df.groupby(['model']).agg(
    avg_mileage = ('mileage','mean')
).reset_index().head()
bdf.sort_values('avg_mileage',ascending = False)
```


What is the average mpg for each model?

```python
df.head()
bdf = df.groupby(['model']).agg(
    avg_mpg = ('mpg','mean')
).reset_index().head()
bdf
```

What is the average price for each fuel type

```python
df.head()
bdf = df.groupby(['fuelType']).agg(
    avg_price= ('price','mean')
).reset_index().head()
bdf
```


What is the average price for each transmission type?

```python
df.head()
bdf= df.groupby(['transmission']).agg(
    Avg_price_trans = ('price','mean')
).reset_index().head()
bdf
bdf.sort_values('Avg_price_trans',ascending=False)
```

What is the average price for each year?

```python
df.head()
bdf = df.groupby(['year']).agg(
    avg_price_year = ('price','mean')
).reset_index().head()
bdf
```


What is the highest price for each model?
