**Sort the dataset by Units Sold in ascending order
**

```python
df.sort_values('Units Sold')
```

**Sort the dataset by Units Sold in descending order.**

```python
df.sort_values('Units Sold', ascending=False)
```

**Sort the dataset by Price per Unit (lowest to highest).**

```python
df.sort_values('Price per Unit')
```

**Sort the dataset by Price per Unit (highest to lowest).**

```python
df.sort_values('Price per Unit' , ascending = False)
```

**Sort all records by Invoice Date in ascending (chronological) order.**

```python
df.sort_values('Invoice Date')
```

**Sort all records by Invoice Date in descending (latest first) order.**

```python
df.sort_values('Invoice Date', ascending = False)
```

**Sort by Product alphabetically (A–Z).**

```python
df.sort_values('Product')
```

**Sort by Product in reverse alphabetical order (Z–A).**

```python
df.sort_values('Product', ascending= False)
```

**Sort by Sales Method alphabetically.**

```python
df.sort_values('Sales Method')
```

**Sort by City and observe if all rows belong to the same city.**

```python
df.sort_values('City')
```

**#Multi-column Sorting**

**Sort by Product ascending, then by Units Sold descending.**

```python
df.sort_values(['Product', "Units Sold"], ascending=[True, False] )
```

**Sort by Region, then by State, both ascending.**

```python
df.sort_values(['Region','State'])
```

**Sort by Sales Method, then Price per Unit descending.**


```python
df.sort_values(['Sales Method', 'Price per Unit'], ascending=[True ,False] )
```

**Sort by Retailer ascending and Invoice Date descending.**

```python
df.sort_values(['Retailer','Invoice Date'], ascending=[True, False])
```

**Sort by City, Product, and Units Sold.**

```python
df.sort_values(['City', 'Product','Units Sold'])
```

**Sort by State ascending and Price per Unit descending.**

```python
df.sort_values(['State','Price per Unit'], ascending=[True ,False])
```

**Sort by Region, Product, and Sales Method.**

```python
df.sort_values(["Region", 'Product', 'Sales Method'])
```

**Sort by Product, then within each product by highest Units Sold.**

```python
sdf= df.sort_values(['Product', 'Units Sold'] , ascending=[True ,False])

sdf
```

***Sort by Product, then Price per Unit, then Invoice Date.***

```python
df.sort_values(['Product', "Price per Unit", 'Invoice Date'])
```

**Calculate a new column Total Sale = Price per Unit * Units Sold and sort descending.**

```python
df['Total Sale'] = df['Price per Unit'] * df['Units Sold']
df.sort_values('Total Sale', ascending=False)
```

**Filter rows where Sales Method == 'In-store' and sort by Units Sold.**

```python
df.loc[df['Sales Method']== 'In-store'].sort_values('Units Sold')
```

Filter rows where Sales Method == 'Outlet' and sort by Price per Unit.

```python
df.loc[df['Sales Method'] == 'Outlet' ].sort_values('Price per Unit')
```

Sort only rows where Units Sold > 1000 by Product.

```python
bdf = df.loc[df['Units Sold']> 1000 ].sort_values('Product')
bdf
```
