```python
df=pd.read_csv('store sales.csv')
df
```

**Find the total sales per quarter using .sum(axis=0).**

```python
df.iloc[: ,1:5].sum()
```

**Find the maximum sales value per quarter using .max(axis=0).**

```python
df.iloc[: ,1:5].max()
```

Find the minimum sales value per quarter using .min(axis=0).

```python
df.iloc[:,1:5].min()
```


Compute the average (mean) sales per quarter using .mean(axis=0).

```python
df.iloc[:,1:5].mean()
```


Compute the median sales per quarter using .median(axis=0).

```python
df.iloc[:,1:5].median()
```


Find the mode of sales per quarter using .mode(axis=0).

```python
df.iloc[:,1:5].mode()
```


Find how many stores reported data for each quarter using .count(axis=0).

```python
df.iloc[:,1:5].count()
```


Identify which quarter had the highest total sales overall.

```python
df.iloc[:,1:5].max()
```

Calculate the range (max âˆ’ min) for each quarter.

```python
mdf=df.iloc[:,1:5].max()
midf=df.iloc[:,1:5].min()
rdf = pd.DataFrame({
    "Max": mdf,
    "Min": midf,
    "Range": mdf - midf
})
rdf
```
