Level 1 – Basic
Create a UDF to calculate Total Amount as Quantity × Price per Unit.

```python
def total_amount(Quantity,Price_per_Unit):
    return Quantity * Price_per_Unit
df['Total_Amount'] = total_amount(df['Quantity'], df['Price per Unit'])
df
```

Create a UDF that categorizes Age into:

Young (<25)
Adult (25–45)
Senior (>45)

```python
def myf(u):
    if u < 25:
        return 'Young'
    elif 25 <= u <=45:
        return 'Adult'
    else:
        return 'Senior'
df['Age_group'] = df['Age'].apply(myf)
df
```


Create a UDF that converts Product Category into single-letter codes:

Beauty → B
Clothing → C
Electronics → E

```python
def Code (y):
    if y == 'Beauty':
        return 'B'
    elif y == 'Clothing':
        return 'C'
    else:
        return 'E'
df['Category_code'] = df['Product Category'].apply(Code)
df
```

Create a UDF that converts Gender:

Male → 1
Female → 0

```python
def gender (g):
    if g == 'Male':
        return 1
    elif g == 'Female':
        return 0
df['gender_code'] = df['Gender'].apply(gender)
df
```

Create a UDF that classifies orders:

Quantity ≥ 3 → High Order

```python
def classified_order (o):
    if o >=3:
        return 'High Order'
    else:
        return 'Low Order'
df['Order_code'] = df['Quantity'].apply(classified_order)
df
```

Create a UDF that creates price bands:

<100 → Budget
100–300 → Mid Range
300 → Premium

```python
def price_band (p):
    if p <100 :
        return 'Budget'
    if 100 <= p <=300:
        return 'Mid Range'
    else:
        return 'Premium'
df['Price_band'] = df['Total_Amount'].apply(price_band)
df
```


Create a UDF that extracts the numeric part from Customer ID (CUST014 → 14) and returns the number of digits.

```python
def digit_find(d):
    num = d.replace('CUST','')
    return len(num)
df['Digit'] = df['Customer ID'].apply(digit_find)
df
```

Based on Total Amount, create a UDF to assign Purchase Tier:
<500 → Bronze
500–1500 → Silver
1500 → Gold

```python
def Purchase(v):
    if v < 500:
        return ' Bronze'
    elif 500 <= v <= 1500:
        return 'Silver'
    else:
        return "Gold"
df['Product_tier'] = df['Total_Amount'].apply(Purchase)
df
```

Create a UDF to assign discount percentage:
Age < 25 → 10%
Price per Unit > 300 → 15%
Otherwise → 5%

```python
def discounting (row):
    if row['Age'] < 25:
        return '10%'
    elif row['Price per Unit'] > 300:
        return '15%'
    else:
        return '5%'
df['Discount_percentage'] = df.apply(discounting ,axis = 1)
df
```

Create a UDF to calculate Final Bill:
Final Bill = Total – (Total × Discount%)


```python
df['Discount_percentage'] = df['Discount_percentage'].str.replace('%','').astype(float)
def final_bill(row):
    return row['Total_Amount'] - (row['Total_Amount']* row['Discount_percentage']/100)
df['Final Bill'] = df.apply(final_bill, axis=1)
df
```

Create a UDF that generates a composite Insurance Score:
Scoring must use:
Customer_Age
Vehicle_Age
Claim_History (yrs)
Location_Risk
Accidents_Last_3Y

```python
def calculate_risk(row):
    score = 0
    # Driver Age Risk
    if row['Customer_Age'] < 25:
        score += 30
    elif row['Customer_Age'] <= 40:
        score += 20
    else:
        score += 10

    if row['Vehicle_Age'] > 10:
        score += 25
    elif row['Vehicle_Age'] >= 5:
        score += 15
    else:
        score += 5

    if row['Claim_History (yrs)'] > 5:
        score += 25
    elif row['Claim_History (yrs)'] > 0:
        score += 15

    if row['Location_Risk'] == 'High':
        score += 30
    elif row['Location_Risk'] == 'Medium':
        score += 15
    else:
        score += 5

    # Accidents in last 3 years
    score += row['Accidents_Last_3Y'] * 10

    return score

df['risk score'] = df.apply(calculate_risk, axis=1)
df
```

Create a UDF that assigns Final Grade based on score:

```python
def final_grade(score):

    if score >= 80:
        return 'A'
    elif 60 <= score < 80:
        return 'B'
    else:
        return 'C'
df['Final_Grade'] = df['Risk score'].apply(final_grade)
df
```
